```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)
```

```python
xy = np.loadtxt('data/data-01-test-score.csv', delimiter=",", dtype=np.float32)
x_data = xy[:, 0:-1] #행 전체, 열은 첫번째부터 마지막 바로 전까지
y_data = xy[:, [-1]] # 행 전체, 열은 마지막만
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3]) 
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([3, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name= 'bias')
```

```python
hypothesis = tf.matmul(X, W)+b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate= 1e-5)
train = optimizer.minimize(cost)
```

```python
sess =tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: x_data, Y:y_data})
    if step%10 ==0:
        print(step, "Cost :", cost_val, "\nPrediction : \n", hy_val)
```

# Queue Runners

```python
filename_queue = tf.train.string_input_producer(['data/data-01-test-score.csv'],
                                               shuffle=False, name="filename_queue")
```

```python
reader = tf.TextLineReader() #문자열 한줄씩 읽어서 리턴
key, value = reader.read(filename_queue) #value에 읽은 값이 순차적으로 리턴됨
```

```python

#각각의 field에 해당하는 값이 어떤 데이터 타입인지!
#reder에서 읽은 값은 아직 raw 데이터임. 
#데이터를 파싱하는 컴포넌트가 Decoder

#각 필드의 디폴트값 + 데이터 타입인지 지정함. 
# ex) "null"로 했으면, default는 null이고, data type은 string
record_defaults = [[0.], [0.], [0.], [0.]]
# csv
xy = tf.decode_csv(value, record_defaults= record_defaults)
```

```python
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10 )
#펌프질 한번마다 10개씩 가지고 오라는게 batch_size임
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3]) 
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name= 'bias')
```

```python
hypothesis = tf.matmul(X, W)+b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate= 1e-5)
train = optimizer.minimize(cost)
```

```python
sess= tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
# 통상적으로 쓰는 코드
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord= coord)
```

```python
for step in range(2001):
    x_batch, y_batch= sess.run([train_x_batch, train_y_batch]) #펌프질 해서 뽑아오는 것
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {X:x_batch, Y:y_batch})
    if step %10 ==0:
        print(step, "Cost:", cost_val, "Prediction", hy_val)
```

```python
coord.request_stop()
coord.join(threads)
```
